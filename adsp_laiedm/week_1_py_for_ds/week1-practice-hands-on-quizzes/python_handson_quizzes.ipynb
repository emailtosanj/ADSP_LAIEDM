```python
#Data visualization quiz

import pandas as pd

df = pd.read_csv(file_dir+'/prog_book.csv')

df.head()


```

```python
#Book data set
df.describe(include='all').T
```

```python
#get the unique values of the type in df data frame
df.Type.unique()
```

```python
#value counts
df.Type.value_counts()
```

```python
#strip plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))
sns.stripplot(x='Type', y='Rating', data=df, hue='Type')

#how do i change color for each Type in the strip plot
```

```python
#box plot for rating
plt.figure(figsize=(14,10))
sns.boxplot(y='Rating', data=df)
```

```python
#distribution plot for reviews
plt.figure(figsize=(10,5))
sns.distplot(df['Reviews'], kde=True)
```

```python
#
df.describe(include='number').T
```

```python
lst = ['Rating', 'Reviews', 'Number_Of_Pages', 'Price']
corr = df[lst].corr()

plt.figure(figsize = (14,10))
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt= '.1f', xticklabels=corr.columns,
            yticklabels = corr.columns)

# Plot the heatmap
#plt.figure(figsize = (14, 10))
#sns.heatmap(corr, annot = True, cmap = 'coolwarm',
#        fmt = ".1f",
#        xticklabels = corr.columns,
#        yticklabels = corr.columns)
```
